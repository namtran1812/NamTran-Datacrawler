What is the role of the test set in the design of end-to-end machine learning projects? How is it incorporated during the training stage?

The test set is crucial in the design of end-to-end machine learning projects. Its role is to provide an unbiased estimate of the performance of the model on new, unseen data. The test set, which serves a distinct purpose from that of the training set used to build the model and the validation set used in tuning hyperparameters and handling overfitting, is generally used once. You test the model on the test set, hence giving you a realistic view of how the model should perform with real-life information, and if it would generalize outside the data set you have trained it with.

The test set will be independent of the training process. As a matter of fact, the test set should be completely blind in training and validation to avoid any leakage of information. Basically, this separation ensures fairness and accuracy in the performance evaluation. In the training phase, the model is being trained on the training set and tested on a different set called the validation set. It may be used to tune hyperparameters, select the best model to avoid overfitting, and allow the model to optimize without biasing the test set. Only in cases where training is finished, hyperparameter tuning is complete, and the model has been finalized should the test set be employed for evaluation. This would constitute the real measure of performance of a model and the capability of generalization on new data.

Suppose you have 100 training samples that you are using to train a classifier to distinguish between four classes. The training data has 50 samples of class 1, 25 samples of class 2, 20 samples of class 3 and 5 samples of class 4. To evaluate the stability and performance of your classifier on each class, you use 10-fold cross-validation. Is it a good strategy to randomly partition the data into 10 folds? Why or why not? If yes, fully justify why. If no, state why not, provide an alternate cross-validation scheme and justify the new scheme.

It is not a good strategy to randomly partition the data into 10 folds. There are 50 samples in Class 1, 25 in Class 2, 20 in Class 3, and only 5 in Class 4. IThere are 50 samples in Class 1, 25 in Class 2, 20 in Class 3, and only 5 in Class 4. Therefore, if we were to divide randomly, some folds may end up without enough representatives for minority class (Class 4). Random partitioning would likely result into some folds that do not contain any sample from class 4. This means that the classifier would never learn about that minority class during those folds and thus its performance for this particular class could not be evaluated as such. Another problem related to this is that there may be no representation of minority class in certain folds. Such a situation would imply that general performance metrics like accuracy, precision, recall and F1 score do not reflect the real performance of the classifier across all classes. Hence one might draw a wrong conclusion regarding how well a given classifier performs on each individual class especially considering minority one.

An alternate cross-validation scheme I would like to propose is stratified k-fold cross-validation. The problem of class imbalance is addressed by stratified k-fold cross-validation, as it ensures that each fold maintains the same ratio of each class just like the whole dataset. This type of validation guarantees that every single fold comprises a representative sample from all existing classes, even those less frequent among them. In this way, it eliminates situations where a few folds do not have samples from some categories and gives an unbiased assessment of how well classifiers perform on different categories in combination. Thus stratified k-fold helps improve accuracy and precision about model’s capacity for dealing with all classes particularly rare ones whose numbers are very small.

# Exercise 8 (5 points)

What are the two most common approaches for encoding categorical attributes? Provide at least 1 advantage and 1 challenge for each.

The two most common approaches for encoding categorical attributes are one-hot encoding and label encoding.

1. One-hot enconding:
* Advantage: One-hot encoding is capable of managing categorical variables in an efficient way as it eliminates possible ordinal relations that do not exist among categories. It makes sure that no category signifies a higher or lower value than another one; hence, preserving the integrity of data.
* Challenge: One-hot encoding can give rise to a large dimensionality of features cube as far as numerous unique categories are considered including dealing with categorical attributes. This may cause the model's computational costs, time complexity and other problems related to dimensions like oversimplification which occurs when classifier fails to fit training data set and is forced into generalization rather than differentiating instances from one another.

2. Label encoding:
* Advantage: Label encoding is simple and computationally inexpensive as it converts categories into one integer column. This method works particularly well with categorical variables which have natural ordinality (low, medium, high), where numeric ordering of labels represents always existing ordering in data.
* Challenge: A challenge that comes with label encoding is that it can create an unintentional ordinal relationship among categories which may mislead a model into thinking that there is a significant order or distance among them. Thus, for entirely nominal data lacking any built-in sequence, such a model might get misled about what it means by ordering and one could observe performance degradation.
